```python
import requests
import folium
from folium import plugins
from folium.plugins import FloatImage
```

```python
titulo = 'Mapa Regulariza'
titulo_html = '''
             <h3 align="center" style="font-size:20px"><b>{}</b></h3>
             '''.format(titulo)
```

```python
mapa_regulariza = folium.Map(location=[-13.29, -39.66], zoom_start=11, control_scale=True)
folium.TileLayer(tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
                 attr= 'Tiles &copy; Esri &mdash; Source: Esri, i-cubed, USDA, USGS, AEX, GeoEye, Getmapping, Aerogrid, IGN, IGP, UPR-EGP, and the GIS User Community',
                 name= 'Esri.WorldImagery').add_to(mapa_regulariza)

mapa_regulariza.get_root().html.add_child(folium.Element(titulo_html))

url = ('https://raw.githubusercontent.com/cleitoncajueiro/Mapa_Clientes_Regulariza/main/Simbolos/Rosa%20dos%20Ventos.png')
FloatImage(url, bottom=0, left=90, width = '10%',).add_to(mapa_regulariza)

geojson_data = requests.get(
    "https://raw.githubusercontent.com/cleitoncajueiro/Mapa_Clientes_Regulariza/main/Clientes_Regulariza.geojson"
).json()

folium.GeoJson(geojson_data, name="Clientes Regulariza", tooltip = 'Clientes Regulariza',).add_to(mapa_regulariza)

folium.LayerControl().add_to(mapa_regulariza)
```

```python
mapa_regulariza.add_child(folium.LatLngPopup())
```

```python
mapa_regulariza.save('index.html')
```
